```python
#ALPHA
#FOR TRANSPARENCY OF A LINE
#IN ZERO THE LINE WILL DISAPPEAR
plt.plot([1,2,3,4],[5,6,7,8],marker="+",markeredgecolor="red",linewidth=30,alpha=0.21)#THE COLOR OF LINE BECOME LIGHTER
plt.show()
plt.plot([1,2,3,4],[5,6,7,8],marker="+",markeredgecolor="red",linewidth=30,alpha=10)
plt.show()
```

```python
#ZORDER
#TO SET THAT WHICH LINE BECOME IN FRONT AND WHICH LINE GOEAS TO BACK
plt.plot([1,2,3,4],[5,6,7,8],marker="+",markeredgecolor="red",linewidth=30)
plt.plot([1,2,3,4],[5,6,9,10],marker="+",markeredgecolor="green",linewidth=30)
plt.show()
#BY DEAULT LAST PLOT WILL BECOME IN FRONT
#BUT NOW IF WE WANT BLUE LINE IN FRONT SO WE WILL GIVE LAGER ZORDER VALUE TO IT
plt.plot([1,2,3,4],[5,6,7,8],marker="+",markeredgecolor="red",linewidth=30,zorder=2)
plt.plot([1,2,3,4],[5,6,9,10],marker="+",markeredgecolor="red",linewidth=30,zorder=1)
plt.show()

plt.plot([1,2,3,4],[5,6,7,8],marker="+",markeredgecolor="red",linewidth=30,zorder=2)
plt.plot([1,2,3,4],[5,6,9,10],marker="+",markeredgecolor="green",linewidth=30,zorder=3)#THIS LINE BECAME IN FRONT BEACAUSE OF LARGER ZORDER VALUE
plt.plot([1,2,3,4],[5,6,11,12],marker="+",markeredgecolor="black",linewidth=30,zorder=1)
plt.show()
```

## **LEGEND**

```python
#LEGEND USED TO DESCRIBE THE PLOT
#IT WILL GIVE INFORMATION ABOUT PLOT
#TO USE LEGEND, WE GIVE LABEL TO EVERY PLOTTING DATA
plt.plot([1,2,3],[4,5,6],Label="FIRST PLOT")
plt.plot([7,8,9],[10,11,12],Label="SECOND PLOT")
plt.title("FIRST PLOT")
plt.xlabel("X AXIS")
plt.ylabel("Y AXIS")
plt.legend()
plt.show()
#THE BOX IS CALLED LEGEND AND THE THINGS PRESENT IN LEGEND IS CALLED LEGEND ENTRIES
#GREEN LINE AND ORANGE LINE IS CALLED KEY AND FIRST PLOT AND SECOND PLOT IS CALLED VALUES
```

```python
#TO CHANGE THE LOCATION OF THE LEGEND
#BY DEFAULT IS BEST
# LOC STRING             LOC CODE
# "BEST"                    0
#  UPPER RIGHT              1
#  UPPER LEFT               2
#  LOWER LEFT               3
#  LOWER RIGHT              4
#  RIGHT                    5
#  CENTER LEFT              6
#  CENTER RIGHT             7
#  LOWER CENTER             8
#  UPPER CENTER             9
#  CENTER                  10
plt.plot([4,6,9,10],label="first")
plt.plot([5,12,-5],label="second")
plt.legend(loc="center")
plt.show()


plt.plot([4,6,9,10],label="first")
plt.plot([5,12,-5],label="second")
plt.legend(loc="right")
plt.show()

plt.plot([4,6,9,10],label="first")
plt.plot([5,12,-5],label="second")
plt.legend(loc="upper center")
plt.show()

#OR
#WE CAN USE LOC CODE TO GIVE THE LOCATION
plt.plot([4,6,9,10],label="first")
plt.plot([5,12,-5],label="second")
plt.legend(loc=3)
plt.show()

plt.plot([4,6,9,10],label="first")
plt.plot([5,12,-5],label="second")
plt.legend(loc=8)
plt.show()

plt.plot([4,6,9,10],label="first")
plt.plot([5,12,-5],label="second")
plt.legend(loc=1)
plt.show()
```

```python
#NCOL
#TO GIVE THE NUMBER OF COLUMN BYDEFAULT IS 1
plt.plot([4,6,9,10],label="first")
plt.plot([5,12,-5],label="second")
plt.legend(ncol=2,loc="center")
plt.show()

plt.plot([4,6,9,10],label="first")
plt.plot([5,12,-5],label="second")
plt.plot([4,6,9,10],label="third")
plt.plot([5,12,-5],label="fourth")
plt.legend(ncol=4)
plt.show()
```

```python
#TIGHT_LAYOUT
#TO DISPLAY EVERY THING EVEN WHICH IS PRESENT OUTSIDE THE PLOT. THE USE OF THIS FUNCTION SEE IN IDLE NOT HERE
#LOC=((X VALUE,Y VALUE))=TO DISPLAY LEGEND OUTSIDE THE PLOT
#FOR EXAMPLE
#LOC=((1,1))
#LOC=((1.5,1))
#LOC=((1,1.5))
#LOC=((1,2))
plt.plot([4,6,9,10],label="first")
plt.plot([5,12,-5],label="second")
plt.plot([4,6,9,10],label="third")
plt.plot([5,12,-5],label="fourth")
plt.legend(loc=(1,1))
plt.tight_layout()
plt.show()

plt.plot([4,6,9,10],label="first")
plt.plot([5,12,-5],label="second")
plt.plot([4,6,9,10],label="third")
plt.plot([5,12,-5],label="fourth")
plt.legend(loc=(1,1.5))
plt.tight_layout()
plt.show()


plt.plot([4,6,9,10],label="first")
plt.plot([5,12,-5],label="second")
plt.plot([4,6,9,10],label="third")
plt.plot([5,12,-5],label="fourth")
plt.legend(loc=(1,0))
plt.tight_layout()
plt.show()

plt.plot([4,6,9,10],label="first")
plt.plot([5,12,-5],label="second")
plt.plot([4,6,9,10],label="third")
plt.plot([5,12,-5],label="fourth")
plt.legend(loc=(2,1))
plt.tight_layout()
plt.show()
```
